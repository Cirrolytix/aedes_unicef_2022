```
import requests
import pandas as pd
import io
```

```
cities = {
    'CAVITE': '14.3294,120.9367',
    'LAGUNA': '14.2167,121.1667',
    'CEBU': '10.3,123.9',
    'ILOILO': '10.7167,122.5667',
    'QUEZON_CITY': '14.6333,121.0333',
    'PANGASINAN': '15.9281,120.3489',
    'BULACAN': '14.7928,120.8789',
    'NEGROS_OCCIDENTAL': '9.9833,122.8167',
    'BATANGAS': '13.75,121.05',
    'RIZAL': '14.5842,121.1763',
    'PAMPANGA': '15.0333,120.6833',
    'BUKIDNON': '7.9,125.0833',
    'NUEVA_ECIJA': '15.4833,120.9667',
    'TARLAC': '15.4802,120.5979',
    'NEGROS ORIENTAL': '9.3103,123.3081',
    'BOHOL': '9.65,123.85',
    'SOUTH_COTABATO': '6.5,124.85',
    'QUEZON': '7.7333,125.1',
    'ZAMBOANGA_CITY': '6.9167,122.0833',
}

params = 'TS,T2M,QV2M,RH2M,T2MDEW,T2MWET,GWETTOP,T2M_MAX,T2M_MIN,GWETPROF,GWETROOT,CLOUD_AMT,T2M_RANGE,PRECTOTCORR,ALLSKY_SFC_LW_DWN'

start = 1985
end = 2021
```

```
def get_weather(city, latlong , start, end, params):
    latitude, longitude = latlong.split(",")
    
    req_url  = 'https://power.larc.nasa.gov/api/temporal/daily/point?' 

    req_query  = {
        'parameters': params,
        'community': 'ag',
        'latitude' : latitude, 
        'longitude' : longitude,
        'start': start, 
        'end' : end, 
        'format' : 'csv',
        'header' : 'false'
        }    

    data = requests.get(req_url, params=req_query).content.decode('utf8')
    df = pd.read_csv(io.StringIO(data))
    df['loc'] = city
    print('Scraping done for: ', city)
    return df
```

```
new_df = pd.DataFrame()
for city, latlong in cities.items():
    if(len(new_df) == 0):
        new_df = get_weather(city, latlong, start, end, params)
    else:
        new_df = new_df.append(get_weather(city, latlong, start, end, params))
```

```
new_df.to_csv('city_weather.csv')
```
