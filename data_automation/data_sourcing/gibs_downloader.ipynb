```
#key: layername, value1: layer, value2: wrap
layers = {
    # 'combined_fapar' : 'MODIS_Combined_L4_FPAR_4Day',
    'terra_fapar': 'MODIS_Terra_L4_FPAR_8Day',
    'ndvi' : 'MODIS_Terra_L3_NDVI_16Day',
    # 'soil_temp' : 'SMAP_L4_Soil_Temperature_Layer_1',
    # 'edvi' : 'MODIS_Terra_L3_EVI_16Day',
    # 'surface_relative_humidity_day': 'AIRS_L2_Surface_Relative_Humidity_Day',
    # 'precipitation_day_est' : 'AIRS_Precipitation_Day',
    # 'precipitation_night_est' : 'AIRS_Precipitation_Night',
    # 'land_surface_temp_day' : 'MODIS_Terra_L3_Land_Surface_Temp_Daily_Day_TES', 
    # 'land_surface_temp_night' : 'MODIS_Terra_L3_Land_Surface_Temp_Daily_Night_TES',
}

cities = {
    'CAVITE': '14.3294,120.9367',
    'ZAMBOANGA_CITY': '6.9167,122.0833',
    'LAGUNA': '14.2167,121.1667',
    'CEBU': '10.3,123.9',
    'ILOILO': '10.7167,122.5667',
    'QUEZON_CITY': '14.6333,121.0333',
    'PANGASINAN': '15.9281,120.3489',
    'BULACAN': '14.7928,120.8789',
    'NEGROS_OCCIDENTAL': '9.9833,122.8167',
    'BATANGAS': '13.75,121.05',
    'RIZAL': '14.5842,121.1763',
    'PAMPANGA': '15.0333,120.6833',
    'BUKIDNON': '7.9,125.0833',
    'NUEVA_ECIJA': '15.4833,120.9667',
    'TARLAC': '15.4802,120.5979',
    'NEGROS ORIENTAL': '9.3103,123.3081',
    'BOHOL': '9.65,123.85',
    'SOUTH_COTABATO': '6.5,124.85',
    'QUEZON': '7.7333,125.1',
}

def latlong_gen(city, diff):
    latlong = cities[city]
    lat_v, long_v = latlong.split(",")
    
    #define size of BBOX?
    LL_lat = round(float(lat_v) - diff,4)
    LL_long = round(float(long_v) - diff,4)
    UR_lat = round(float(lat_v) + diff, 4)
    UR_long = round(float(long_v) + diff, 4)

    bbox = '"' + str(LL_lat) + "," + str(LL_long) + '" "' + str(UR_lat) + "," + str(UR_long) + '"' 
    
    return bbox
```

```
# [0.03, 0.06, 0.125, 0.25, 0.5, 1, 5, 10]
resolution = 0.03 

keep_xml = "false"
download_faster = "true" # multiprocessing of python

start_date = "2013-01-01"
end_date = "2022-08-30"
size_of_bbox = 0.11

for city_name, city_latlong in cities.items():
    for layer_name, layer_product in layers.items():
        
        city_bbox = latlong_gen(city_name, size_of_bbox)
        sat_product = layer_product
        output_path = "images/" + city_name + "/" + layer_name

        if not os.path.exists(output_path):
            os.makedirs(output_path)

        !gdl {start_date} {end_date} {city_bbox} \
         --name={sat_product} --name={sat_product} \
         --res={resolution} --mp={download_faster} \
         --output-path={output_path} \
         --keep-xml={keep_xml}
```
